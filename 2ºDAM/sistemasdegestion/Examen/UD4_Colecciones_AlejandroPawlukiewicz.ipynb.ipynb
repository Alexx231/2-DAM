```python
# EJERCICIO 1: lISTAS DE PRODUCTOS

# Lista inicial de productos
productos = [
    ["Manzana", 1.2, 50],
    ["Banana", 0.5, 100],
    ["Melón", 3.0, 20],
    ["Mango", 2.5, 30],
    ["Naranja", 1.0, 80]
]

# 1. Añadir un nuevo producto
nuevo_producto = ["Pera", 1.5, 60]
productos.append(nuevo_producto)
print("Lista con el nuevo producto añadido:")
for producto in productos:
    print(producto)

print("---------------------")

# 2. Modificar el precio del segundo producto
productos[1][1] = 0.75
print("Lista con el precio del segundo producto modificado:")
for producto in productos:
    print(producto)

print("---------------------")

# 3. Ordenar los productos en función de su precio de forma descendente
productos.sort(key=lambda x: x[1], reverse=True)
print("Lista de productos ordenados por precio de forma descendente:")
for producto in productos:
    print(producto)

print("---------------------")

# 4. Filtrar los productos cuyo stock sea mayor de 20 unidades utilizando comprensión de listas
productos_stock_mayor_20 = [producto for producto in productos if producto[2] > 20]
print("Lista de productos con stock mayor a 20 unidades:")
for producto in productos_stock_mayor_20:
    print(producto)

print("---------------------")

# 5. Calcular el valor total del inventario (precio * cantidad) y almacenarlo en una nueva lista
valor_inventario = [[producto[0], producto[1] * producto[2]] for producto in productos]
print("Lista con el valor total del inventario para cada producto:")
for producto in valor_inventario:
    print(producto)
```

```python
# EJERCICIO 2: TUPLAS Y VENTAS

# 1. Crear una tupla con los días de la semana en los que la tienda está abierta (de lunes a sábado)
dias_abiertos = ("Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado")

# 2. Crear un conjunto de ventas registradas en los días "Lunes", "Martes" y "Miércoles"
ventas_registradas = {"Lunes", "Martes", "Miércoles"}
print("Conjunto de ventas registradas:", ventas_registradas)

# 3. Intentar modificar un día en la tupla y explicar por qué no es posible
# dias_abiertos[0] = "Domingo"  # Esto generará un error porque las tuplas son inmutables
# Comentario: No es posible modificar un elemento de una tupla porque las tuplas son inmutables en Python.

# 4. Convertir la tupla a una lista, añadir el domingo como un día laborable, y volver a convertirla en tupla
dias_abiertos_lista = list(dias_abiertos)
dias_abiertos_lista.append("Domingo")
dias_abiertos = tuple(dias_abiertos_lista)
print("Tupla con el domingo añadido:", dias_abiertos)

# 5. Escribir una función que reciba como parámetro un día de la semana y devuelva si la tienda está abierta o cerrada ese día
def esta_abierta(dia):
    if dia in dias_abiertos:
        return "La tienda está abierta."
    else:
        return "La tienda está cerrada."

# Ejemplos de uso de la función
print(esta_abierta("Lunes"))  # La tienda está abierta.
print(esta_abierta("Domingo"))  # La tienda está abierta.
```

```python
# EJERCICIO 3: GESTION DE EMPLEADOS CON CONJUNTOS

# Conjuntos iniciales de empleados
ventas = {"Ana", "Luis", "Carlos", "Marta"}
marketing = {"Marta", "Pedro", "Luis", "Sofía"}

# 1. Mostrar los empleados que trabajan en ambos departamentos
empleados_ambos = ventas.intersection(marketing)
print("Empleados que trabajan en ambos departamentos:")
print(empleados_ambos)

print("---------------------")

# 2. Añadir un nuevo empleado a ventas
ventas.add("Jorge")
print("Lista de empleados en ventas después de añadir un nuevo empleado:")
print(ventas)

print("---------------------")

# 3. Eliminar un empleado de marketing
marketing.remove("Pedro")
print("Lista de empleados en marketing después de eliminar un empleado:")
print(marketing)

print("---------------------")

# 4. Encontrar los empleados que solo trabajan en marketing
solo_marketing = marketing.difference(ventas)
print("Empleados que solo trabajan en marketing:")
print(solo_marketing)

print("---------------------")

# 5. Calcular la cantidad total de empleados únicos en la empresa
empleados_unicos = ventas.union(marketing)
print("Cantidad total de empleados únicos en la empresa:")
print(len(empleados_unicos))
```

```python
# EJERCICIO 4: GESTION DE INVENTARIO CON DICCIONARIOS

# Diccionario inicial del inventario
inventario = {
    "Leche": {"precio": 1.9, "stock": 50},
    "Cereal": {"precio": 2.0, "stock": 10},
    "Yogur": {"precio": 1.5, "stock": 20},
    "Carne": {"precio": 2.7, "stock": 50},
    "Verdura": {"precio": 1.2, "stock": 75}
}

# 1. Añadir un nuevo producto al inventario
inventario["Proteina"] = {"precio": 1.1, "stock": 60}
print("Inventario después de añadir un nuevo producto:")
for producto in inventario:
    print(producto)

print("---------------------")

# 2. Modificar el precio de un producto
producto_a_modificar = "Cereal"
if producto_a_modificar in inventario:
    inventario[producto_a_modificar]["precio"] = 0.75
    print(f"Precio de {producto_a_modificar} modificado a 0.75")
else:
    print(f"El producto {producto_a_modificar} no existe en el inventario.")

print("Inventario después de modificar el precio de un producto:")
for producto, datos in inventario.items():
    print(f"Producto: {producto}, Precio: {datos['precio']}, Stock: {datos['stock']}")


print("---------------------")

# 3. Reducir el stock de un producto después de una venta
def vender_producto(nombre, cantidad):
    if nombre in inventario:
        if inventario[nombre]["stock"] >= cantidad:
            inventario[nombre]["stock"] -= cantidad
            print(f"Se han vendido {cantidad} unidades de {nombre}.")
        else:
            print(f"No hay suficiente stock de {nombre}.")
    else:
        print(f"El producto {nombre} no existe en el inventario.")

producto_a_vender = "Carne"
cantidad_a_vender = 10
vender_producto(producto_a_vender, cantidad_a_vender)

print("Inventario después de una venta:")
for producto, datos in inventario.items():
    print(f"Producto: {producto}, Precio: {datos['precio']}, Stock: {datos['stock']}")

print("---------------------")

# 4. Crear una función que reciba como parámetro el nombre de un producto y devuelva su precio y cantidad en stock
def obtener_info_producto(nombre):
    if nombre in inventario:
        return inventario[nombre]
    else:
        return f"El producto {nombre} no existe en el inventario."

info_producto = obtener_info_producto("Verdura")
print("Información del producto Verdura:")
print(info_producto)

print("---------------------")

# 5. Generar un reporte de todos los productos en el inventario, mostrando el valor total del inventario
def generar_reporte(inventario):
    valor_total = 0
    print("Reporte del inventario:")
    for producto, datos in inventario.items():
        valor_producto = datos["precio"] * datos["stock"]
        valor_total += valor_producto
        print(f"Producto: {producto}, Precio: {datos['precio']}, Stock: {datos['stock']}, Valor Total: {valor_producto}")
    print(f"Valor total del inventario: {valor_total}")

generar_reporte(inventario)
```
